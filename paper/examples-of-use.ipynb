```python
from symqups.simple import *
import sympy
from IPython.display import display
```

## EXAMPLE 1

```python
omega, lmbda, Gamma, n_mean = sympy.symbols(r"omega, lambda, Gamma, n_\mathrm{avg}")
H = hbar.val*omega*adOp*aOp + lmbda*adOp + dagger(lmbda*adOp)
dissip_1 = [2*Gamma*(1+n_mean), aOp]
dissip_2 = [2*Gamma*n_mean, adOp]
lme = LindbladMasterEquation(H, dissip_1, dissip_2)

s.val = s.default_value
result = collect_by_derivative(sympy.expand(CGTransform(lme)))
display(result)
print(sp.latex(result))
```

```python
display(result.subs({Gamma : 0, lmbda : 0}))
print(sp.latex(result.subs({Gamma : 0, lmbda : 0})))
```

```python
s.val = -1
display(collect_by_derivative(sympy.expand(CGTransform(lme))).subs(Gamma, 0))
print(sp.latex(collect_by_derivative(sympy.expand(CGTransform(lme))).subs(Gamma, 0)))
```

```python
s.val = 1
display(collect_by_derivative(sympy.expand(CGTransform(lme))).subs({omega:0, lmbda:0}))
print(sp.latex(collect_by_derivative(sympy.expand(CGTransform(lme))).subs({omega:0, lmbda:0})))
```

## EXAMPLE 2

```python
kappa = sp.symbols("kappa", real=True)

hbar.val = 1
s.val = 0

H = kappa/2 * adOp**2 * aOp**2

lme = LindbladMasterEquation(H)

display(collect_by_derivative(sp.expand(CGTransform(lme))))
print(sp.latex(collect_by_derivative(sp.expand(CGTransform(lme)))))
```

## EXAMPLE 3

```python
kappa_1, gamma_1, gamma_2 = sp.symbols("kappa_1, gamma_1, gamma_2", real=True)

hbar.val = 1
s.val = 0

H = adOp*aOp + sp.Rational(1,2)
dissip_1 = [kappa_1, adOp]
dissip_2 = [gamma_1, aOp]
dissip_3 = [gamma_2, aOp**2]

lme = LindbladMasterEquation(H, dissip_1, dissip_2, dissip_3)

display(collect_by_derivative(sp.expand(CGTransform(lme))))
print(sp.latex(collect_by_derivative(sp.expand(CGTransform(lme)))))
```
